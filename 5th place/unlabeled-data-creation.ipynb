```python
import pandas as pd
import numpy as np
import glob
from tqdm.auto import tqdm
import os
```

```python
INPUT_DIR = '/kaggle/input/tlvmc-parkinsons-freezing-gait-prediction'
OUTPUT_DIR = './'
```

```python
os.mkdir(f'{OUTPUT_DIR}/unlabeled_np')
```

```python
pq_files = glob.glob(f'{INPUT_DIR}/unlabeled/*.parquet')
```

```python
metadata = []
```

```python
CHUNK_SIZE = 100000
```

```python
for fname in tqdm(pq_files):
    data = pd.read_parquet(fname)
    base_name = os.path.basename(fname).replace('.parquet','')
    for ch in range(len(data)//CHUNK_SIZE):
        arr = data.loc[ch*CHUNK_SIZE:(ch+1)*CHUNK_SIZE]
        file_name = f'{OUTPUT_DIR}/unlabeled_np/{base_name}_chunk_{ch}.npz'
        np.savez_compressed(file_name,arr=arr)
        
        metadata.append({'base_name':base_name,'file_name':file_name})
    #break
```

```python
metadata = pd.DataFrame(metadata)
metadata.to_csv('{OUTPUT_DIR}/unlabeled_metadata.csv',index=False)
```
